```python
# Marketing Analytics Portfolio - Python Visualizations
# Created by: Ashlyn Talley
# Dataset: Facebook Ads Performance Analysis (951 campaigns)
# Customized for your actual data structure

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.patches import Rectangle

# Set professional styling
plt.style.use('default')  # Using default for better compatibility
sns.set_palette("Set2")

print("🎯 Marketing Analytics Portfolio - Data Visualizations")
print("=" * 60)

# Your data is already loaded as 'df'
print(f"📊 Dataset loaded: {len(df)} campaigns analyzed")
print(f"📈 Performance distribution: {df['performance_label'].value_counts().to_dict()}")

# ====================
# 1. VISUALIZATION 1: Performance Distribution & Key Metrics
# ====================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Pie chart - Performance Label Distribution
performance_counts = df['performance_label'].value_counts()
colors = ['#DC143C', '#2E8B57', '#FFD700']  # Red for low, Green for high, Gold for medium
labels = ['Low Performance', 'High Performance', 'Medium Performance']

# Reorder to match your data (low=805, high=101, medium=45)
ordered_counts = [performance_counts['low'], performance_counts['high'], performance_counts['medium']]
wedges, texts, autotexts = ax1.pie(ordered_counts,
                                   labels=labels,
                                   autopct='%1.1f%%',
                                   colors=colors,
                                   startangle=90,
                                   explode=(0, 0.05, 0))  # Highlight high performers

ax1.set_title('Campaign Performance Distribution\n(n=951 campaigns)',
              fontsize=14, fontweight='bold', pad=20)

# Make percentage text bold and readable
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)

# Bar chart - Performance Metrics by Label
performance_metrics = df.groupby('performance_label').agg({
    'ctr': 'mean',
    'roas': 'mean',
    'cpa': 'mean'
}).round(2)

# Reorder for logical presentation
performance_metrics = performance_metrics.reindex(['low', 'medium', 'high'])

x = np.arange(len(performance_metrics.index))
width = 0.25

bars1 = ax2.bar(x - width, performance_metrics['ctr'], width,
                label='CTR (%)', color='#4CAF50', alpha=0.8)
bars2 = ax2.bar(x, performance_metrics['roas'], width,
                label='ROAS (x)', color='#2196F3', alpha=0.8)
bars3 = ax2.bar(x + width, performance_metrics['cpa']/10, width,
                label='CPA ($, ÷10)', color='#FF9800', alpha=0.8)

ax2.set_xlabel('Performance Tier', fontweight='bold')
ax2.set_ylabel('Metric Value', fontweight='bold')
ax2.set_title('Average Metrics by Performance Tier', fontweight='bold', pad=15)
ax2.set_xticks(x)
ax2.set_xticklabels(['Low', 'Medium', 'High'])
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

# Add value labels on bars
for bars in [bars1, bars2, bars3]:
    for bar in bars:
        height = bar.get_height()
        ax2.annotate(f'{height:.1f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('performance_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Chart 1: Performance Distribution - Shows clear performance differentiation")

# ====================
# 2. VISUALIZATION 2: Feature Correlation Analysis
# ====================

plt.figure(figsize=(10, 8))

# Select your actual marketing metrics for correlation
metrics = ['ctr', 'cpa', 'roas', 'spend', 'impressions', 'clicks', 'conversions', 'cpm']
correlation_matrix = df[metrics].corr()

# Create heatmap
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
heatmap = sns.heatmap(correlation_matrix,
                      mask=mask,
                      annot=True,
                      cmap='RdYlBu_r',
                      center=0,
                      square=True,
                      fmt='.2f',
                      cbar_kws={"shrink": .8},
                      annot_kws={'fontsize': 10, 'fontweight': 'bold'})

plt.title('Facebook Ads Metrics Correlation Analysis\nUnderstanding Feature Relationships for ML Training',
          fontsize=16, fontweight='bold', pad=20)

# Customize labels
labels = ['CTR (%)', 'CPA ($)', 'ROAS (x)', 'Spend ($)', 'Impressions', 'Clicks', 'Conversions', 'CPM ($)']
heatmap.set_xticklabels(labels, rotation=45, ha='right', fontweight='bold')
heatmap.set_yticklabels(labels, rotation=0, fontweight='bold')

plt.tight_layout()
plt.savefig('correlation_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Chart 2: Correlation Analysis - Shows feature relationships for ML model training")

# ====================
# 3. VISUALIZATION 3: Performance Insights Dashboard
# ====================

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Facebook Ads Performance Analysis Dashboard\nML Training Data Insights & Business Intelligence',
             fontsize=18, fontweight='bold', y=0.98)

# Subplot 1: CTR vs ROAS Scatter Plot with Performance Labels
high_perf = df[df['performance_label'] == 'high']
med_perf = df[df['performance_label'] == 'medium']
low_perf = df[df['performance_label'] == 'low']

ax1.scatter(low_perf['ctr'], low_perf['roas'], alpha=0.6, c='#DC143C',
           label=f'Low ({len(low_perf)})', s=50)
ax1.scatter(med_perf['ctr'], med_perf['roas'], alpha=0.6, c='#FFD700',
           label=f'Medium ({len(med_perf)})', s=50)
ax1.scatter(high_perf['ctr'], high_perf['roas'], alpha=0.6, c='#2E8B57',
           label=f'High ({len(high_perf)})', s=50)

ax1.set_xlabel('Click-Through Rate (%)', fontweight='bold')
ax1.set_ylabel('Return on Ad Spend (x)', fontweight='bold')
ax1.set_title('CTR vs ROAS by Performance Tier\n(Clear Feature Separation)', fontweight='bold')
ax1.legend()
ax1.grid(alpha=0.3)

# Subplot 2: Spend Efficiency Analysis
ax2.scatter(df['spend'], df['conversions'], alpha=0.6, c=df['cpa'],
           cmap='RdYlGn_r', s=50)  # Reversed colormap so green = low CPA
colorbar = plt.colorbar(ax2.collections[0], ax=ax2)
colorbar.set_label('Cost Per Acquisition ($)', fontweight='bold')

ax2.set_xlabel('Ad Spend ($)', fontweight='bold')
ax2.set_ylabel('Conversions', fontweight='bold')
ax2.set_title('Spend vs Conversions (Color = CPA)', fontweight='bold')
ax2.grid(alpha=0.3)

# Subplot 3: Performance Metrics Distribution
metrics_data = [df[df['performance_label'] == label]['roas'] for label in ['low', 'medium', 'high']]
box_plot = ax3.boxplot(metrics_data, labels=['Low', 'Medium', 'High'],
                      patch_artist=True, notch=True)

colors = ['#DC143C', '#FFD700', '#2E8B57']
for patch, color in zip(box_plot['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

ax3.set_xlabel('Performance Label', fontweight='bold')
ax3.set_ylabel('Return on Ad Spend (ROAS)', fontweight='bold')
ax3.set_title('ROAS Distribution by Performance Tier\n(ML Target Variable Analysis)', fontweight='bold')
ax3.grid(axis='y', alpha=0.3)

# Subplot 4: Key Performance Indicators Table
ax4.axis('off')
ax4.set_title('Facebook Ads Dataset - Key Insights', fontweight='bold', pad=20)

# Calculate key statistics
total_spend = df['spend'].sum()
total_conversions = df['conversions'].sum()
avg_ctr = df['ctr'].mean()
avg_roas = df['roas'].mean()
avg_cpa = df['cpa'].mean()

# High performer insights
high_avg_ctr = high_perf['ctr'].mean()
high_avg_roas = high_perf['roas'].mean()
high_avg_cpa = high_perf['cpa'].mean()

kpi_data = [
    ['📊 Total Campaigns', f'{len(df):,}'],
    ['💰 Total Ad Spend', f'${total_spend:,.0f}'],
    ['🎯 Total Conversions', f'{total_conversions:,.0f}'],
    ['📈 Average CTR', f'{avg_ctr:.2f}%'],
    ['🚀 Average ROAS', f'{avg_roas:.2f}x'],
    ['💳 Average CPA', f'${avg_cpa:.2f}'],
    ['⭐ High Performer CTR', f'{high_avg_ctr:.2f}%'],
    ['⭐ High Performer ROAS', f'{high_avg_roas:.2f}x'],
    ['⭐ High Performer CPA', f'${high_avg_cpa:.2f}']
]

table = ax4.table(cellText=kpi_data,
                 colLabels=['Metric', 'Value'],
                 cellLoc='left',
                 loc='center',
                 colWidths=[0.6, 0.4])

table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 1.8)

# Style the table
for i in range(len(kpi_data) + 1):
    for j in range(2):
        cell = table[(i, j)]
        if i == 0:  # Header row
            cell.set_facecolor('#4CAF50')
            cell.set_text_props(weight='bold', color='white')
        elif i > 6:  # High performer rows
            cell.set_facecolor('#E8F5E8')
            cell.set_text_props(weight='bold' if j == 1 else 'normal')
        else:
            cell.set_facecolor('#f9f9f9' if i % 2 == 0 else 'white')
            cell.set_text_props(weight='bold' if j == 1 else 'normal')

plt.tight_layout()
plt.savefig('performance_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Chart 3: Comprehensive Performance Dashboard - Business Intelligence Ready")

# ====================
# 4. INSIGHTS SUMMARY FOR PORTFOLIO
# ====================

print("\n" + "="*60)
print("🎯 KEY INSIGHTS FROM FACEBOOK ADS ANALYSIS")
print("="*60)

# Calculate improvement metrics
ctr_improvement = (high_perf['ctr'].mean() / low_perf['ctr'].mean() - 1) * 100
roas_improvement = (high_perf['roas'].mean() / low_perf['roas'].mean() - 1) * 100
cpa_reduction = (1 - high_perf['cpa'].mean() / low_perf['cpa'].mean()) * 100

print(f"📊 Dataset Quality & Annotation:")
print(f"   • {len(df)} Facebook ad campaigns successfully annotated")
print(f"   • Distribution: {len(low_perf)} Low | {len(med_perf)} Medium | {len(high_perf)} High")
print(f"   • Clear performance tiers with measurable differences")

print(f"\n🚀 Performance Analysis Insights:")
print(f"   • High performers achieve {ctr_improvement:.1f}% higher CTR ({high_perf['ctr'].mean():.2f}% vs {low_perf['ctr'].mean():.2f}%)")
print(f"   • High performers deliver {roas_improvement:.1f}% better ROAS ({high_perf['roas'].mean():.2f}x vs {low_perf['roas'].mean():.2f}x)")
print(f"   • High performers reduce CPA by {cpa_reduction:.1f}% (${high_perf['cpa'].mean():.2f} vs ${low_perf['cpa'].mean():.2f})")

print(f"\n💡 ML Training Data Quality:")
print(f"   • Rich feature set: CTR, CPA, ROAS, Spend, Impressions, Clicks, Conversions, CPM")
print(f"   • Clear target variable with {len(high_perf)} positive examples")
print(f"   • Strong feature separation visible in correlation analysis")
print(f"   • Real-world Facebook Ads Manager data (not synthetic)")

print(f"\n🎯 Business Recommendations:")
print(f"   • Target CTR > {high_perf['ctr'].quantile(0.25):.2f}% for optimal performance")
print(f"   • Aim for ROAS > {high_perf['roas'].quantile(0.25):.2f}x to achieve high performance")
print(f"   • Monitor CPA to stay below ${high_perf['cpa'].quantile(0.75):.2f} for efficiency")

print(f"\n📈 Portfolio Value Proposition:")
print(f"   • Processed and annotated 951 real Facebook ad campaigns")
print(f"   • Created ML-ready training dataset with clear performance labels")
print(f"   • Demonstrated strong analytical skills with Python visualizations")
print(f"   • Extracted actionable business insights from marketing data")

print("\n✅ Analysis Complete! Professional charts saved for GitHub portfolio.")
print("📁 Files created: performance_distribution.png, correlation_analysis.png, performance_dashboard.png")
print("🚀 Ready for AI Trainer job application!")
```

```python
# IMPROVED VERSION - Fix the scaling issue in the bar chart
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# ====================
# IMPROVED CHART 1: Better Performance Distribution Visualization
# ====================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Left: Pie chart (this was fine)
performance_counts = df['performance_label'].value_counts()
colors = ['#DC143C', '#2E8B57', '#FFD700']  # Red, Green, Gold
labels = ['Low Performance', 'High Performance', 'Medium Performance']

ordered_counts = [performance_counts['low'], performance_counts['high'], performance_counts['medium']]
wedges, texts, autotexts = ax1.pie(ordered_counts,
                                   labels=labels,
                                   autopct='%1.1f%%',
                                   colors=colors,
                                   startangle=90,
                                   explode=(0, 0.05, 0))

ax1.set_title('Campaign Performance Distribution\n(n=951 campaigns)',
              fontsize=14, fontweight='bold', pad=20)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)

# Right: IMPROVED - Separate charts for different metrics to solve scaling
performance_metrics = df.groupby('performance_label').agg({
    'ctr': 'mean',
    'roas': 'mean',
    'cpa': 'mean'
}).round(2)

performance_metrics = performance_metrics.reindex(['low', 'medium', 'high'])

# Create 3 separate subplots within the right panel
gs = ax2.get_gridspec()
ax2.remove()

# Create 3 smaller subplots
ax2a = fig.add_subplot(gs[0, 1])  # This will be split into 3
ax2b = plt.subplot2grid((3, 2), (0, 1), fig=fig)
ax2c = plt.subplot2grid((3, 2), (1, 1), fig=fig)
ax2d = plt.subplot2grid((3, 2), (2, 1), fig=fig)

# CTR Chart
bars1 = ax2b.bar(['Low', 'Medium', 'High'], performance_metrics['ctr'],
                color=['#DC143C', '#FFD700', '#2E8B57'], alpha=0.8)
ax2b.set_ylabel('CTR (%)', fontweight='bold', fontsize=10)
ax2b.set_title('Click-Through Rate', fontweight='bold', fontsize=11)
ax2b.grid(axis='y', alpha=0.3)

# Add value labels
for bar in bars1:
    height = bar.get_height()
    ax2b.annotate(f'{height:.1f}%',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom',
                fontsize=9, fontweight='bold')

# ROAS Chart
bars2 = ax2c.bar(['Low', 'Medium', 'High'], performance_metrics['roas'],
                color=['#DC143C', '#FFD700', '#2E8B57'], alpha=0.8)
ax2c.set_ylabel('ROAS (x)', fontweight='bold', fontsize=10)
ax2c.set_title('Return on Ad Spend', fontweight='bold', fontsize=11)
ax2c.grid(axis='y', alpha=0.3)

# Add value labels
for bar in bars2:
    height = bar.get_height()
    ax2c.annotate(f'{height:.1f}x',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom',
                fontsize=9, fontweight='bold')

# CPA Chart
bars3 = ax2d.bar(['Low', 'Medium', 'High'], performance_metrics['cpa'],
                color=['#DC143C', '#FFD700', '#2E8B57'], alpha=0.8)
ax2d.set_ylabel('CPA ($)', fontweight='bold', fontsize=10)
ax2d.set_xlabel('Performance Tier', fontweight='bold', fontsize=10)
ax2d.set_title('Cost Per Acquisition', fontweight='bold', fontsize=11)
ax2d.grid(axis='y', alpha=0.3)

# Add value labels
for bar in bars3:
    height = bar.get_height()
    ax2d.annotate(f'${height:.1f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom',
                fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('improved_performance_distribution.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Improved Chart 1: Now all metrics are clearly visible!")

# ====================
# ALTERNATIVE VERSION - Side-by-side comparison table
# ====================

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Pie chart (same as before)
wedges, texts, autotexts = ax1.pie(ordered_counts,
                                   labels=labels,
                                   autopct='%1.1f%%',
                                   colors=colors,
                                   startangle=90,
                                   explode=(0, 0.05, 0))

ax1.set_title('Campaign Performance Distribution\n(n=951 campaigns)',
              fontsize=14, fontweight='bold', pad=20)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(11)

# Table showing clear metrics comparison
ax2.axis('off')
ax2.set_title('Performance Metrics Comparison', fontweight='bold', fontsize=14, pad=20)

# Create comparison table
table_data = []
metrics_info = [
    ('CTR (%)', performance_metrics['ctr'], '%'),
    ('ROAS (x)', performance_metrics['roas'], 'x'),
    ('CPA ($)', performance_metrics['cpa'], '$')
]

for metric_name, values, unit in metrics_info:
    if unit == '$':
        row = [metric_name,
               f"${values['low']:.1f}",
               f"${values['medium']:.1f}",
               f"${values['high']:.1f}"]
    elif unit == '%':
        row = [metric_name,
               f"{values['low']:.1f}%",
               f"{values['medium']:.1f}%",
               f"{values['high']:.1f}%"]
    else:
        row = [metric_name,
               f"{values['low']:.1f}{unit}",
               f"{values['medium']:.1f}{unit}",
               f"{values['high']:.1f}{unit}"]
    table_data.append(row)

# Add improvement calculations
ctr_improvement = ((performance_metrics.loc['high', 'ctr'] / performance_metrics.loc['low', 'ctr']) - 1) * 100
roas_improvement = ((performance_metrics.loc['high', 'roas'] / performance_metrics.loc['low', 'roas']) - 1) * 100
cpa_reduction = (1 - (performance_metrics.loc['high', 'cpa'] / performance_metrics.loc['low', 'cpa'])) * 100

table_data.append(['', '', '', ''])  # Spacer row
table_data.append(['High vs Low:', '', '', ''])
table_data.append(['CTR Improvement', f"+{ctr_improvement:.0f}%", '', ''])
table_data.append(['ROAS Improvement', f"+{roas_improvement:.0f}%", '', ''])
table_data.append(['CPA Reduction', f"-{cpa_reduction:.0f}%", '', ''])

table = ax2.table(cellText=table_data,
                 colLabels=['Metric', 'Low', 'Medium', 'High'],
                 cellLoc='center',
                 loc='center',
                 colWidths=[0.3, 0.2, 0.2, 0.2])

table.auto_set_font_size(False)
table.set_fontsize(11)
table.scale(1, 2)

# Style the table
colors_row = ['#DC143C', '#FFD700', '#2E8B57']
for i in range(len(table_data) + 1):
    for j in range(4):
        cell = table[(i, j)]
        if i == 0:  # Header row
            cell.set_facecolor('#4CAF50')
            cell.set_text_props(weight='bold', color='white')
        elif i <= 3:  # Metric rows
            if j == 0:  # Metric name column
                cell.set_facecolor('#f0f0f0')
                cell.set_text_props(weight='bold')
            elif j > 0:  # Data columns
                cell.set_facecolor('#f9f9f9')
                cell.set_text_props(weight='bold')
        elif i > 4:  # Improvement rows
            if j <= 1:
                cell.set_facecolor('#E8F5E8')
                cell.set_text_props(weight='bold')
            else:
                cell.set_facecolor('white')

plt.tight_layout()
plt.savefig('performance_comparison_table.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Alternative Chart: Clear table format showing all metrics!")
print(f"\n📊 Key Insights Now Clearly Visible:")
print(f"   • High performers have {ctr_improvement:.0f}% better CTR")
print(f"   • High performers achieve {roas_improvement:.0f}% better ROAS")
print(f"   • High performers reduce CPA by {cpa_reduction:.0f}%")
```
